### Получаем массив квалификаций для каждого работника

```python
quals = []
for row in Qualified.index:
    counter = []
    for column in Qualified:
        if Qualified[column][row] == 1:
            counter.append(column)
    if len(counter)==0:
        counter.append('non')
    quals.append(counter)
```

```python
quals = pd.DataFrame(quals, columns = ['quals_of_worker'])
```

### Добавляем этот массив к основной информации о работнике

```python
info_about_worker = pd.concat([info_about_worker, quals], axis=1)
```

### Получаем массив запрошенных отпусков для каждого работника, затем их приоритеты

```python
restsreq = []
for row in RestReq.index:
    counter = []
    for column in RestReq:
        if RestReq[column][row] != 0:
            counter.append(column)
    if len(counter)==0:
        counter.append('non')
    restsreq.append([counter])
```

```python
restsreq = pd.DataFrame(restsreq, columns = ['rests_req_of_worker'])
```

```python
info_about_worker = pd.concat([info_about_worker, restsreq], axis = 1)
info_about_worker
```

```python
# Добавим количество часов в запросах
restsreqhours = []
for row in RestReq.index:
    counter = []
    for column in RestReq:
        if RestReq[column][row] != 0:
            counter.append(RestReq[column][row])
    if len(counter)==0:
        counter.append('non')
    restsreqhours.append([counter])
```

```python
restsreqhours = pd.DataFrame(restsreqhours, columns=['rests_req_hours_of_worker'])
restsreqhours
```

```python
info_about_worker = pd.concat([info_about_worker, restsreqhours], axis = 1)
info_about_worker
```

```python
# Добавляем приоритеты
restsprior = []
for row in RestPrior.index:
    counter = []
    for column in RestPrior:
        if RestPrior[column][row] != 0:
            counter.append(RestPrior[column][row])
    if len(counter)==0:
        counter.append('non')
    restsprior.append([counter])
```

```python
restsprior = pd.DataFrame(restsprior, columns=['rests_priors_of_worker'])
restsprior
```
